```python
import os

# download the “texts.zip” file, extract it, and upload all .txt files.
# this is the same as uploading .csv files before

# Get your current working directory
cwd = os.getcwd()
```

```python
corpus = []
# reading files from the directory and adding them to a list
for subdir, dirs, files in os.walk(cwd):
    for file in files:
        if file.endswith(".txt"):
            f=open(os.path.join(subdir, file),'r', encoding="utf8")
            a = f.read()
            corpus.append(a)
            f.close()
```

```python
corpus[0]
```

```python
import re # re module provides regular expression matching operations
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
# first document before any modification and in its original format
print(corpus[0])
```

**Preprocess Text**

```python
StopWords = stopwords.words("english")

# basic operations on text data
for i in range(len(corpus)):
    doc = corpus[i]

    # substitutes newline/tab/space with one space in the text
    doc = re.sub('\s+', ' ', doc)

    # remove numbers
    doc = re.sub(r'[0-9]+', '', doc)  

    # remove punctuation
    doc = re.sub(r'[^\w\s]', '', doc) 

    # convert to lower case
    doc = doc.lower() 

    # remove stop words
    doc = ' '.join([word for word in doc.split() if word not in StopWords])

    corpus[i] = doc

# first document after applying all of the above operations
# You can check that there is no stop word, no ponctuation, and all letters are lowercase now
print(corpus[0])
```

**Extract most frequent words in each document**

```python
for i in range(len(corpus)):
    doc = corpus[i]

    # get the list of all the words in the string
    split_it = doc.split()

    # find most frequent words in document (after preprocessing)
    from collections import Counter
    Counter = Counter(split_it)
    # get top-5 most frequent words
    most_freq = Counter.most_common(5)

    print('most frequent words in document', (i+1), ' is:', most_freq)
```

```python
# show a word cloud of the first document

from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                min_font_size = 10).generate(corpus[0])

# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
